```python
# Expense Sharing App (Fixed Version)

users = []
expenses = []

# Add user
def add_user(name):
    name = name.strip()
    users.append(name)
    print(f"{name} added successfully!")

# Equal split
def split_equal(amount, participants):
    share = round(amount / len(participants), 2)
    return {user: share for user in participants}

# Custom split
def split_custom(participants):
    shares = {}
    total = 0
    print("Enter share for each user:")
    
    for user in participants:
        val = float(input(f"{user}: "))
        shares[user] = val
        total += val

    return shares

# Add expense
def add_expense():
    payer = input("Paid by: ").strip()
    amount = float(input("Amount: "))
    
    participants = input("Split among (comma separated): ").split(',')
    participants = [p.strip() for p in participants]  # ✅ FIX

    split_type = input("Split type (equal/custom): ").strip().lower()

    if split_type == "equal":
        shares = split_equal(amount, participants)
    else:
        shares = split_custom(participants)

    expense = {
        "payer": payer,
        "amount": amount,
        "participants": participants,
        "shares": shares
    }

    expenses.append(expense)
    print("Expense added successfully!")

# Show balances
def show_balances():
    print("\n--- Balances ---")

    balance = {user: 0 for user in users}

    for exp in expenses:
        payer = exp["payer"]

        for user in exp["participants"]:
            share = exp["shares"][user]

            if user == payer:
                balance[user] += exp["amount"] - share
            else:
                balance[user] -= share

    for user, amt in balance.items():
        print(f"{user}: {round(amt, 2)}")

# Menu
while True:
    print("\n--- Expense Sharing App ---")
    print("1. Add User")
    print("2. Add Expense")
    print("3. Show Balances")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == '1':
        name = input("Enter user name: ")
        add_user(name)

    elif choice == '2':
        add_expense()

    elif choice == '3':
        show_balances()

    elif choice == '4':
        print("Thank you!")
        break

    else:
        print("Invalid choice!")
```
